```python
import pandas as pd
import urllib.parse
```

```python
df = pd.read_json("dane-egipt.json")
df.head()
```

```python
df.columns
```

```python
columns = ["id","name","link","category","duration","durationNights","departureDate","returnDate","ratingValue","price","priceDiscount","serviceDesc"]
```

```python
df = df[columns]
df.drop_duplicates(inplace=True)
df
```

```python
def decode_link(s):
  _ , url = s.split("::")
  return "http://wakacje.pl" + urllib.parse.unquote(url)[::-1]
```

```python
df_result = df.query(" serviceDesc=='All Inclusive' and ratingValue>8 and departureDate>='2024-03-01' and departureDate<'2024-04-01' and category>=4 ") \
      .sort_values(["price","ratingValue","category"], ascending=[True,False,False])
df_result.link = df_result.link.apply(decode_link)
df_result
```
